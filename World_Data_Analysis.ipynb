```python
import pandas as pd

def read_worldbank_data(filename):
    # Read the data into a DataFrame
    df = pd.read_csv(filename, skiprows=4)

    # Transpose the DataFrame and set the index
    df_t = df.transpose()
    df_t.columns = df_t.iloc[0]
    df_t = df_t.iloc[1:]

    # Clean up the DataFrame
    # df_t.index.name = 'Year'
    # df_t.columns.name = 'Country'
    # df_t = df_t.apply(pd.to_numeric, errors='coerce')

    # Create a DataFrame with years as columns
    df_year = df_t.transpose()

    # Create a DataFrame with countries as columns
    df_country = df_t.copy()

    return df_year, df_country
```

```python
#Formed a Function
def plot_indicator_by_country(df, indicator_code, countries, title, ylable, xlabel):
    # create an empty dataframe to store the results
    df_tmp = pd.DataFrame()

    # loop over each country and append the data to the dataframe
    for country in countries:
        # select the rows for the country and the indicator of interest
        df_country = df.loc[(df['Country Code'] == country) & (df['Indicator Code'] == indicator_code)]
        # append the data for this country to the main dataframe
        df_tmp = pd.concat([df_tmp, df_country], axis=0)

    # select the desired years of the dataframe
    df_tmp = df_tmp.loc[:, '2018':'2021']

    # plot the data
    ax = df_tmp.plot(kind='bar', figsize=(16, 6), lw=2, colormap='Set1')

    # set the title and axis labels
    ax.set_title(title, fontsize=16)
    ax.set_xlabel(xlabel, fontsize=14)
    ax.set_ylabel(ylabel, fontsize=14)

    # set the legend
    ax.legend(title='Country', fontsize=12, title_fontsize=14, bbox_to_anchor=(1.02, 1))

    # show the plot
    plt.show()
```

```python
def plot_timeseries_by_country(df, indicator_code, countries, title, ylabel, xlabel):
    # create an empty dataframe to store the results
    df_tmp = pd.DataFrame()

    # loop over each country and append the data to the dataframe
    for country in countries:
        # select the rows for the country and the indicator of interest
        df_country = df.loc[(df['Country Code'] == country) & (df['Indicator Code'] == indicator_code)]
        # append the data for this country to the main dataframe
        df_tmp = pd.concat([df_tmp, df_country], axis=0)

    # select the desired years of the dataframe
    df_tmp = df_tmp.loc[:, '1960':'2021']

    # set the index to be the countries and transpose the dataframe
    df_tmp = df_tmp.set_index('Country Name').T

    # plot the data
    ax = df_tmp.plot(kind='line', figsize=(16, 6), lw=2, colormap='Set1')

    # set the title and axis labels
    ax.set_title(title, fontsize=16)
    ax.set_xlabel(xlabel, fontsize=14)
    ax.set_ylabel(ylabel, fontsize=14)

    # set the legend
    ax.legend(title='Country', fontsize=12, title_fontsize=14, bbox_to_anchor=(1.02, 1))

    # show the plot
    plt.show()
```
